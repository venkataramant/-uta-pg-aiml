```python
import pandas as pd
```

```python
loc_example=pd.DataFrame(data={"city":["Hyderabad","Khammam","Warangal"],
              "population":[10000000,300000,300000],
              "distirct":["Rangareddy","Sambhadri","Orugallu"]

             },
                         index=["id1","id2","id3"])
loc_example
```

```python
# loc is indexer takes argument in labels - index in string(if string otherwise int), column names in string
print(loc_example.loc["id1"]) # Fetch entire row  with index=id1
print("\nchoose only selected columns\n")
print(loc_example.loc["id1",["city"]]) # choose only selected columns
```

```python
# iloc is indexer takes argument in integer - for both index/columns
print(loc_example.iloc[0]) # Fetch entire row  with index=id1
print("\nchoose only selected columns\n")
print(loc_example.iloc[[0,1],[0,2]]) # choose only selected columns
```

```python
## if index is interger , use interger in loc
loc_example2=pd.DataFrame(data={"city":["Hyderabad","Khammam","Warangal"],
              "population":[10000000,300000,300000],
              "distirct":["Rangareddy","Sambhadri","Orugallu"],

             })
loc_example2
# loc row
print(loc_example2.loc[0])
print("\nchoose only selected columns\n")
print(loc_example2.loc[1,["city"]])
print("\npass multiple values, choose only selected columns\n")
print(loc_example2.loc[[0,1],["population","city"]]) # Takes one or list of elements
# Error - print(loc_example2.loc[[0:],["city"]]) - Slicing is not allowed
```

```python
df1=pd.DataFrame({"city":["Hyderabad","Khammam","Warangal"],
              "population":[10000000,300000,300000],

             })
df1
```

```python
df2=pd.DataFrame({"city":["Vizag","Guntur","Vijaywada"],
              "population":[30000000,300000,1000000],

             })
df2.loc[len(df2)]=['Nellore',500000]
df2
```

```python
## Concat by default stacks one df on top of another one when axis=0
## when axis=1, keep dfs side by side (Columnns are duplicated)
df3=pd.concat([df1,df2])
print(df3)
# To reset index values use "ignore_index"
print("\nignore_index=True\n")
df3=pd.concat([df1,df2],ignore_index=True)
print(df3)

print("\n Use keys additonal indexing\n")
df4=pd.concat([df1,df2],keys=["TG","AP"])
print(df4)
```

```python
print("\n Get one cell\n")
print(df4.loc["TG"].loc[0,"city"])
print("\naccess 2 rows\n")
print(df4.loc["TG"].loc[[0,2],"city"])
```

```python
## Concat by default stacks one df on top of another one
# axis=1, merges dfs side by side using index_value, if any index_value is not there those cells have Nan
df3=pd.concat([df1,df2],axis=1)
df3
```

```python
### join - by default outer, "inner" chooses records if index value is matched.
pd.concat([df1,df2],axis=1,join="inner")
```

```python
pd.concat([df1,df2],axis=1,join="outer") # outer is default value
```
