```python
import pandas as pd
df1=pd.DataFrame({
    "city": ["Hyderabad","Vijayawada","Warangal"],
    "humidity" :(40,50,60),
"temperature" :(25,30,27)
})
df1
```

```python
df2=pd.DataFrame({
    "city": ["Hyderabad","Khammam","Vizag"],
"humidity" :(60,40,80)
})
df2
```

```python
# merge will work as join on SQL side
# on, how(inner/outer/right/left),suffixes
print(df1)
print()
print(df2)
print()
print("inner_"*5)
df3=pd.merge(df1,df2,on=["city"],how="inner")
print(df3)


print()
print("outer_"*3)
df3=pd.merge(df1,df2,on=["city"],how="outer",indicator=True,suffixes=("_left","_right"))
print(df3)
print()
df3=pd.merge(df1,df2,on=["city"],how="right")
print(df3)
print("right")
print()
df3=pd.merge(df1,df2,on=["city"],how="left")
print(df3)
print("left")
print()
```
